```python
"""
we import the libraries.
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, accuracy_score, f1_score, classification_report, confusion_matrix, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from joblib import Memory
from imblearn.over_sampling import SMOTE
import seaborn as sns
from pandas import get_dummies
from sklearn.preprocessing import StandardScaler
#
```

## 1. Data Loading

```python
data = pd.read_csv('job_salary_prediction_dataset.csv', nrows=5000)
X = data.drop(columns=['salary'])
Y = data['salary']
```

## 2. Data Preparation

```python
#checking if there are null values in the dataset
data.isnull().sum()
data.drop_duplicates(inplace=True)
```

We split the data into training and testing sets.

```python
#we need to numerize the data
X = get_dummies(X)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)   
```

## 3. Model Creation

We can use Linear Regression

```python
Model = LinearRegression()
Model.fit(x_train, y_train) 
y_pred = Model.predict(x_test)
```

## 4. Metrics Evaluation

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("mae : ", mae)
print("rmse : ", rmse)
print("r2 : ", r2)
```
